# Age column cleaning and creating new column with age groups

```python
# Changing age type to integer.
# Defining age groups in the function age_grouping()

def age_to_int(x):
    if str(x).isdigit():
        return int(x)
    else:
        return "Unknown"

def age_groupping(x):
    if isinstance(x, int):
        if x <= 10:
            return "Below 10"

        elif x > 10 and x <= 20:
            return "10-20"
        elif x > 20 and x <= 30:
            return "20-30"
        elif x > 30 and x <= 40:
            return "30-40"
        elif x > 40 and x <= 50:
            return "40-50"
        elif x > 50 and x <= 60:
            return "50-60"
        elif x > 60:
            return "Above 60"

    return "Unknown"
```

```python
df_20['age']=df_20['age'].apply(age_to_int)

df_20['age_groups'] = df_20['age'].apply(age_groupping)
df_20.age_groups.value_counts()
```

# Standardizing the column activity and grouping activities

```python
# Standardizing the column 'activity' with consistent values & removing any additional spaces

df_20['activity'] = df_20['activity'].str.strip()

other_fishing = "spearfishing"
fishing = "fishing"

df_20['activity'] = df_20['activity'].replace(other_fishing, fishing)

other_swimming = ['wading', 'bathing', 'treading water', 'swimming']
swimming = "swimming"

df_20['activity'] = df_20['activity'].replace(other_swimming, swimming)

other_diving = ['scuba diving', 'free diving', 'pearl diving']
diving = "diving"

df_20['activity'] = df_20['activity'].replace(other_diving, diving)

other_surfing = ['standing', 'body boarding', 'body surfing', 'boogie boarding', 'fell overboard', 'windsurfing']
surfing = "surfing"

df_20['activity'] = df_20['activity'].replace(other_surfing, surfing)
```

```python
# changing the data type of the column 'activity' to string type
df_20['activity'] = df_20['activity'].astype(str)

# converting each value of the 'activity' column to lowercase
df_20['activity'] = df_20['activity'].str.lower()
```

```python
# defining a function to group similar type of activities in one single category (example: spearfishing in fishing)

def group_act(x):

    surfing = "surf"
    swimming = "swim"
    fishing = "fish"
    diving = "diving"
    snorkeling = "snorkeling"
    kayaking = "kayak"

    if surfing in x:
        return "surfing"
    if swimming in x:
        return "swimming"
    if fishing in x:
        return "fishing"
    if diving in x:
        return "diving"
    if snorkeling in x:
        return "snorkeling"
    if kayaking in x:
        return "kayaking"
    else:
        return "others"

df_20["activity_group"] = df_20["activity"].apply(group_act)
```

```python
# now checking the value counts in the 'activity_group' column
df_20['activity_group'].value_counts().head(20)
```

# Grouping injuries into six main groups

```python
df_20['injury']=df_20['injury'].astype(str)
df_20['injury']=df_20['injury'].str.lower()
def group_inj(x):


    leg = "leg"
    arm = "arm"
    hand = "hand"
    foot = "foot"
    feet = "feet"
    fatal = "fatal"
    ankle = "ankle"
    chest = "chest"
    body = "body"
    head = "head"
    stomach = "stomach"
    thigh = "thigh"
    calf = "calf"
    calves = "calves"
    finger = "finger"

    if leg in x:
        return "lower limb"
    if chest in x:
        return "body"
    if stomach in x:
        return "body"
    if body in x:
        return "body"
    if ankle in x:
        return "lower limb"
    if calf in x:
        return "lower limb"
    if calves in x:
        return "lower limb"
    if thigh in x:
        return "lower limb"
    if foot in x:
        return "lower limb"
    if feet in x:
        return "lower limb"
    elif arm in x:
        return "upper limb"
    elif hand in x:
        return "upper limb"
    elif finger in x:
        return "upper limb"
    elif fatal in x:
        return "fatal"
    elif head in x:
        return "head"
    else:
        return "Unknown"
```

```python
df_20["injury_group"] = df_20['injury'].apply(group_inj)
```
